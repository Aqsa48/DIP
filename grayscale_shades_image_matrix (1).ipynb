# Grayscale Shades as Matrices - Step by Step

This notebook shows how to create grayscale images with different shades (0=black, 128=gray, 192=light gray, 255=white).

We scale from 2x2 → 4x4 → 8x8 → 16x16 to show how pixel values map to shades of gray.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Create a 2x2 Grayscale Image (different shades)

```python
# 0 = black, 128 = medium gray, 192 = light gray, 255 = white
image_2x2 = np.array([[0, 128],
                      [192, 255]])

print("2x2 Matrix:\n", image_2x2)

plt.imshow(image_2x2, cmap='gray', vmin=0, vmax=255)
plt.title("2x2 Grayscale (Shades)")
plt.axis('off')
plt.show()
```

## Step 2: Scale to 4x4

```python
image_4x4 = np.repeat(np.repeat(image_2x2, 2, axis=0), 2, axis=1)

print("4x4 Matrix:\n", image_4x4)

plt.imshow(image_4x4, cmap='gray', vmin=0, vmax=255)
plt.title("4x4 Grayscale (Shades)")
plt.axis('off')
plt.show()
```

## Step 3: Scale to 8x8

```python
image_8x8 = np.repeat(np.repeat(image_2x2, 4, axis=0), 4, axis=1)

print("8x8 Matrix:\n", image_8x8)

plt.imshow(image_8x8, cmap='gray', vmin=0, vmax=255)
plt.title("8x8 Grayscale (Shades)")
plt.axis('off')
plt.show()
```

## Step 4: Scale to 16x16

```python
image_16x16 = np.repeat(np.repeat(image_2x2, 8, axis=0), 8, axis=1)

print("16x16 Matrix:\n", image_16x16)

plt.imshow(image_16x16, cmap='gray', vmin=0, vmax=255)
plt.title("16x16 Grayscale (Shades)")
plt.axis('off')
plt.show()
```
