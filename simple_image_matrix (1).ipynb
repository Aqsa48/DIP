# Digital Image as a Matrix - Step by Step Example

This notebook shows how to create simple images (2x2, 4x4, 8x8, 16x16) using NumPy arrays and visualize them with Matplotlib.

We explain each step for beginners.

```python
import numpy as np          # NumPy is used to handle matrices (arrays of numbers).
import matplotlib.pyplot as plt  # Matplotlib is used to display images (visual output).
```

## Step 1: Create a 2x2 Image

```python
# A simple 2x2 image: 0 = Black, 255 = White
image_2x2 = np.array([[0, 255],
                      [255, 0]])

print("2x2 Matrix:\n", image_2x2)  # Show the math (numbers)

plt.imshow(image_2x2, cmap='gray', vmin=0, vmax=255)  # Show as image
plt.title("2x2 Image")
plt.axis('on')
plt.show()
```

## Step 2: Scale up to 4x4

```python
# Expand the image to 4x4 by repeating rows and columns
image_4x4 = np.repeat(np.repeat(image_2x2, 2, axis=0), 2, axis=1)

print("4x4 Matrix:\n", image_4x4)

plt.imshow(image_4x4, cmap='gray', vmin=0, vmax=255)
plt.title("4x4 Image")
plt.axis('off')
plt.show()
```

## Step 3: Scale up to 8x8

```python
# Expand to 8x8 (repeat 4 times)
image_8x8 = np.repeat(np.repeat(image_2x2, 4, axis=0), 4, axis=1)

print("8x8 Matrix:\n", image_8x8)

plt.imshow(image_8x8, cmap='gray', vmin=0, vmax=255)
plt.title("8x8 Image")
plt.axis('off')
plt.show()
```

## Step 4: Scale up to 16x16

```python
# Expand to 16x16 (repeat 8 times)
image_16x16 = np.repeat(np.repeat(image_2x2, 8, axis=0), 8, axis=1)

print("16x16 Matrix:\n", image_16x16)

plt.imshow(image_16x16, cmap='gray', vmin=0, vmax=255)
plt.title("16x16 Image")
plt.axis('off')
plt.show()
```
